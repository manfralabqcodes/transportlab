# Initialize Instruments

```python
Rxy=SR860('Laughlin', 'GPIB1::4::INSTR')
#Rxy=SR830('DIlbert', 'GPIB1::6::INSTR')
```

# Set Parameters for Instruments

```python
Freq_Init = 5
Freq_Final = 1000
stepSize = 1
dataPointNum = int((Freq_Final - Freq_Init - 1)/stepSize)
```

# Set Up Database for Data Collection

```python
db_file_path = os.path.join(os.getcwd(), 'C:\\Users\\Admin\Desktop\\MattMann\\8-3-22.1MM11B4\\CutoffFreq\\ZeroField_2Terminal.db')
initialise_or_create_database_at(db_file_path)
```

# Set Up Live Plot

```python
jobs = bg.BackgroundJobManager()
jobs.new(inspectr.main, db_file_path)
```

# Setup Station and Measurement Parameters

```python
station = qc.Station()
station.add_component(Rxy)

cutoff_freq_exp = load_or_create_experiment(
    experiment_name="cutoff_freq_exp2",
    sample_name="8-3-22.1MM11B4"
)
context_meas = Measurement(exp=cutoff_freq_exp, station=station, name='cutoff_freq_exp2')
context_meas.write_period = 0.1

context_meas.register_parameter(Rxy.frequency)
context_meas.register_parameter(Rxy.X, setpoints=(Rxy.frequency,))
context_meas.register_parameter(Rxy.R, setpoints=(Rxy.frequency,))
context_meas.register_parameter(Rxy.P, setpoints=(Rxy.frequency,))

```

# Measurements

```python
with context_meas.run() as datasaver:

    for set_f in np.linspace(Freq_Init, Freq_Final, dataPointNum):
        Rxy.frequency.set(set_f)
        get_x = Rxy.X.get()
        get_r = Rxy.R.get()
        get_p = Rxy.P.get()
        datasaver.add_result((Rxy.frequency, set_f),
                             (Rxy.X, get_x),
                             (Rxy.R, get_r),
                             (Rxy.P, get_p))
        time.sleep(4)

    dataset = datasaver.dataset
```

```python
plot_dataset(dataset)
```
