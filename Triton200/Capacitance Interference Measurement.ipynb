# Set Parameters for Instruments

```python
#Parameters for SIM Controller
#sim.volt_2()
#sim.volt_2.step=0.001
#sim.volt_2.inter_delay=0.2
waitTime = 1
amp = 100000000
step_size = 0.010
V1_start = 0.4
V1_end = 0
V2_start = 0.3
V2_end = 0
V1_SIMPORT = 4
V2_SIMPORT = 6
path = "C:\\Users\\Admin\\Desktop\\MattMann\\8-3-22.1MM11C5\\Capacitance Correction\\QPC7&3 With QPC17.xlsx"
```

# Run Measurement

```python

#exp = load_or_create_experiment('Capacitance Correction with QPC7&3 With QPC18')
#meas = Measurement()
#meas.register_parameter(rxx2.X)

workbook = openpyxl.Workbook()
sheet = workbook.active 
print(rxx2)
c1 = sheet.cell(row = 1, column = 1) 
c1.value = "Vg1"
c2 = sheet.cell(row = 1, column = 2) 
c2.value = "Vg2"
c3 = sheet.cell(row = 1, column = 3) 
c3.value = "Current"
c4 = sheet.cell(row = 1, column = 4) 
c4.value = "Phase"
curr = 2

for i in np.arange(V1_start, V1_end, -2*step_size):
    sim.set_voltage(V1_SIMPORT, i)
    #Sweep Down
    print("Sweep Down")
    for j in np.arange(V2_start, V2_end, -1*step_size):
        temp1 = sheet.cell(row = curr, column = 1)
        temp2 = sheet.cell(row = curr, column = 2)
        temp3 = sheet.cell(row = curr, column = 3)
        temp4 = sheet.cell(row = curr, column = 4)
        sim.set_voltage(V2_SIMPORT, j)
        time.sleep(waitTime)
        temp1.value = i
        temp2.value = j
        (X, P) = rxx2.get_values('X','P')
        temp3.value = X/amp
        temp4.value = P
        print(X)
        curr = curr + 1
    sim.set_voltage(V1_SIMPORT, i-step_size)
    #Sweep Up
    print("Sweep Up")
    for j in np.arange(V2_end, V2_start, step_size):
        temp1 = sheet.cell(row = curr, column = 1)
        temp2 = sheet.cell(row = curr, column = 2)
        temp3 = sheet.cell(row = curr, column = 3)
        temp4 = sheet.cell(row = curr, column = 4)
        sim.set_voltage(V2_SIMPORT, j)
        time.sleep(waitTime)
        temp1.value = i
        temp2.value = j
        (X, P) = rxx2.get_values('X','P')
        temp3.value = X/amp
        temp4.value = P
        print(X)
        curr = curr + 1
    
workbook.save(filename=path)
```
